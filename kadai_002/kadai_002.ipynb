```python
import MeCab
```

```python
mecab_tagger = MeCab.Tagger()
```

```python
text = main_text
```

```python
print(mecab_tagger.parse(text))
```

```python
node = mecab_tagger.parseToNode(text)
```

```python
vocab_list = []
```

```python
while node:
    word = node.surface
    hinshi = node.feature.split(",")[0]
    if hinshi != "":
        if(not word.isnumeric(), word):
            vocab_list.append(word)
    else:
        pass
    node = node.next
```

```python
print(vocab_list)
```

```python
result_text_list = list()
```

```python
for split_text in vocab_list:
    if split_text not in stopwords_list:
        result_text_list.append(split_text)
```

```python
print(result_text_list)
```
